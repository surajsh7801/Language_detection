```python
import pandas as pd
import numpy as np 
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
import joblib
import warnings
warnings.filterwarnings("ignore")
```

```python
x1=np.array(df['Text'])
y=np.array(df['Language'])
cv=CountVectorizer()
x=cv.fit_transform(x1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42)
```

```python
mb=MultinomialNB()
```

```python
mb.fit(x_train,y_train)
```

```python
y_pred=mb.predict(x_test)
```

```python
print("accuracy_score",accuracy_score(y_pred,y_test)*100)
```

```python
print("test_score",mb.score(x_train,y_train)*100)
print("train_score",mb.score(x_test,y_test)*100)
```

```python
print("Classification_report",classification_report(y_pred,y_test))
```

```python
con=confusion_matrix(y_pred,y_test).astype(int)
sns.heatmap(con,annot=True,fmt='.0f');
```

```python
joblib.dump(mb,'model.pkl')
```

```python
joblib.dump(cv,'vectorizer.pkl')
```
